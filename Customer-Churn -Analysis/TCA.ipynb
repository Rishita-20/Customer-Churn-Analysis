#converted 0 and 1 value of senuior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else :
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
        
```

```python

plt.figure(figsize =(3,3))
ax = sns.countplot(x = "SeniorCitizen",data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python


# 1. Count the occurrences of 'SeniorCitizen' and 'Churn'
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# 2. Calculate percentages of total counts
total_count = len(df)  # Total number of observations
count_data_percentage = (count_data / total_count) * 100

# 3. Plot the stacked bar chart
ax = count_data_percentage.plot(kind='bar', stacked=True, figsize=(6, 4), color=["skyblue", "salmon"], width=0.8)

# 4. Adding title and labels
plt.title('Churn by SeniorCitizen (% of Total)', fontsize=14)
plt.xlabel('SeniorCitizen', fontsize=12)
plt.ylabel('Percentage of Total (%)', fontsize=12)

# 5. Add percentage labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black')

plt.show()
```

```python


# List of the columns you want to create countplots for
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of rows and columns for subplots (adjust based on number of plots)
n_rows = 3  # You can adjust this based on your preference
n_cols = 3  # Adjust the number of columns for the grid

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(17, 17))

# Flatten the axes to make it easier to iterate through
axes = axes.flatten()

# Loop through each column and plot a countplot in each subplot
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i], palette="Set2",hue = df["Churn"])
    axes[i].set_title(f"Count of {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Show the plots
plt.show()
```

#Majority of Customer donot churn tends to have services like PhoneServices,InternetService(DSL) and onlinesecurity enabled.For services like Onlinebackup,techsupport and StreamingTV,churn rates are noticealy higher.
